## Indexing in pandas

The indexing operator and attribute selection are nice because they work just like they do in the rest of the Python ecosystem. As a novice, this makes them easy to pick up and use. However, pandas has its own accessor operators, `loc` and `iloc`. For more advanced operations, these are the ones you're supposed to be using.

### Index-based selection

Pandas indexing works in one of two paradigms. The first is **index-based selection**: selecting data based on its numerical position in the data. `iloc` follows this paradigm.

To select the first row of data in a DataFrame, we may use the following:

```python
health.iloc[0]
```

```python
# Selecting the first row and the second column
first_row_second_column = health.iloc[0, 1]
first_row_second_column
```

### Label-based selection

The second paradigm for attribute selection is the one followed by the `loc` operator: **label-based selection**. In this paradigm, it's the data index value, not its position, which matters.

For example, to get the last entry in `health`, we would now do the following:

```python
health.loc[0, 'age']
```

Both `loc` and `iloc` are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.

This means that it's marginally easier to retrieve rows, and marginally harder to get retrieve columns. To get a column with `iloc`, we can do the following:

```python
health.iloc[:, 1]
```

```python
health.loc[:, 'age']
```

```python
# health.age

health['age']
```

### Manipulating the index

Label-based selection derives its power from the labels in the index. Critically, the index we use is not immutable. We can manipulate the index in any way we see fit.

The `set_index()` method can be used to do the job. Here is what happens when we `set_index` to the `number` field:

```python
health.head()
```

```python
health.set_index("id")
```

So far we've been indexing various strides of data, using structural properties of the DataFrame itself. To do *interesting* things with the data, however, we often need to ask questions based on conditions. 

```python
health.is_male_over_75 == True
```

```python
health.loc[health.is_male_over_75 == True]
```

This cell demonstrates another method for selecting data based on a condition using the `.loc[]` accessor in Pandas. The line `health.loc[health.is_male_over_75 == True]` is used to filter and display rows from the health DataFrame where the is_male_over_75 column is True.

```python
health.loc[(health.sex == 'Male') & (health['age'] > 75)]
```

This cell showcases a more explicit approach to filtering data in a DataFrame using the `.loc[]` accessor in Pandas.

Using `.loc[] `in this way allows for more granular control over the data selection process and is particularly useful in scenarios where multiple conditions need to be checked simultaneously. It's a common practice in data analysis when you need to analyze specific segments or groups within your dataset.

## Introduction to `groupby()`

The `groupby()` method is used to group data in a DataFrame based on one or more columns. This operation involves splitting the data into groups based on some criteria, applying a function to each group independently, and then combining the results.


```python
health.groupby('age').age.count()
```

```python
health.groupby(['dataset']).age.agg([len, min, max])
```

This cell demonstrates the use of the `groupby()` method combined with aggregation functions in Pandas. The line of code `health.groupby(['dataset']).age.agg([len, min, max])` performs the following operations:

**Grouping by 'dataset'** : The `.groupby(['dataset'])` function groups the data in the health DataFrame based on unique values in the 'dataset' column. This is useful for analyzing data separately within each group.

**Aggregating Age Data** : The `.agg([len, min, max])` part applies multiple aggregation functions to the 'age' column of each group. Specifically, it calculates:

* len: The number of entries (or count) in each group.
* min: The minimum age in each group.
* max: The maximum age in each group.

This command is particularly useful for getting a quick statistical overview of numerical data within each category of a categorical column (in this case, 'dataset'). It helps in understanding the distribution and range of values like age across different datasets or groups within the data.

## Descriptive Statistics for 'Age'

Descriptive statistics summarize the central tendency, dispersion, and shape of a datasets distribution. We'll calculate various descriptive statistics for the 'age' column, such as mean, standard deviation, variance, mode, minimum, maximum, and count.


```python
# Descriptive Statistics for 'age'
age_mean = health['age'].mean()
age_std = health['age'].std()
age_variance = health['age'].var()
age_mode = health['age'].mode()[0]  # mode() returns a Series
age_min = health['age'].min()
age_max = health['age'].max()
age_count = health['age'].count()

# # Display the statistics
# (age_mean, age_std, age_variance, age_mode, age_min, age_max, age_count)

print(f"Descriptive Statistics for 'Age':\n")
print(f"Mean Age: {age_mean:.2f}")
print(f"Standard Deviation: {age_std:.2f}")
print(f"Variance: {age_variance:.2f}")
print(f"Mode: {age_mode}")
print(f"Minimum Age: {age_min}")
print(f"Maximum Age: {age_max}")
print(f"Count: {age_count}")

```
